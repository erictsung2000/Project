## 讓所有不買的都變成 Never buy 統一整理

```python
# 將購買項目中包含 "Never" 的所有情況都統一更改為 "Never buy"
df.loc[df["購買項目"].str.strip().str.lower().isin(["never", "never buy any", "nothing"]), "購買項目"] = "Never buy"

# 將購買項目中 "Cold drinks;Never" 的 "Never" 修改為 "Cold drinks"
df.loc[df["購買項目"].str.strip().str.lower() == "cold drinks;never", "購買項目"] = "Cold drinks"

print(df["購買項目"].unique())
```

```python
df["購買項目"].unique()
```

```python
df["年收入"].unique()
```

```python
def income_to_value(income_str):
    if income_str == "Less than RM25,000":
        return 0
    elif income_str == "RM25,000 - RM50,000":
        return 1
    elif income_str == "RM50,000 - RM100,000":
        return 2
    elif income_str == "RM100,000 - RM150,000":
        return 3
    elif income_str == "More than RM150,000":
        return 4
    else:
        return -1  

# 將收入轉換為數值格式
df["年收入數值"] = df["年收入"].apply(income_to_value)
print(df["年收入數值"])
```

```python
income_groups = df.groupby("年收入數值")

income_groups_product_combination = income_groups["購買項目"].agg(lambda x: ", ".join(x))
income_groups_avg_spend = income_groups["花費金額"].mean()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 8))

sns.countplot(data=df, x='年收入數值', hue='購買項目')

plt.title('不同收入群的顧客購買產品組合分布')
plt.xlabel('年收入群')
plt.ylabel('購買產品組合數量')

plt.legend(loc='upper right',fontsize = "x-small")

plt.show()

plt.figure(figsize=(12, 8))

sns.barplot(data=df, x='年收入數值', y='花費金額', estimator=np.mean)

plt.title('不同收入群的顧客平均花費金額分布')
plt.xlabel('年收入群')
plt.ylabel('平均花費金額')

plt.show()
```

## 圖表得知
**<span style='color: blue; font-size: 20px'>收入More than RM150,000 的消費能力比較高大多都是買 Coffee ，可能是因為星巴克的價格其實並不便宜，真的要收入一般的人每天去消費，可能大家意願不高，可能可以使用會員卡的方式讓大家累積點數，然後再去兌換免費一杯 Coffee ，這樣大家可能會覺得他可能剩一兩點就可以集齊一杯免費的 Coffee，就會想要趕快的去做消費。</span>**

## 問題6 女性和男性在拜訪頻率、滯留時間、花費金額、選擇評價等方面有什麼差異?

```python
gender_groups = df.groupby("性別")

gender_freq = gender_groups["拜訪頻率"].value_counts()
gender_duration = gender_groups["滯留時間"].value_counts()
gender_spend = gender_groups["花費金額"].mean()
gender_choice = gender_groups["選擇評價"].mean()
```

```python
# 創建4個子圖
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

gender_freq.plot(kind='bar', ax=axes[0, 0], title='不同性別的拜訪頻率分布')
axes[0, 0].set_xlabel('拜訪頻率')
axes[0, 0].set_ylabel('數量')

gender_duration.plot(kind='bar', ax=axes[0, 1], title='不同性別的滯留時間分布')
axes[0, 1].set_xlabel('滯留時間')
axes[0, 1].set_ylabel('數量')

gender_spend.plot(kind='bar', ax=axes[1, 0], title='不同性別的平均花費金額')
axes[1, 0].set_xlabel('性別')
axes[1, 0].set_ylabel('平均花費金額')

gender_choice.plot(kind='bar', ax=axes[1, 1], title='不同性別的選擇評價均值')
axes[1, 1].set_xlabel('性別')
axes[1, 1].set_ylabel('選擇評價均值')

plt.tight_layout()

plt.show()
```

## 圖中表示
**<span style='color: blue; font-size: 20px'>拜訪的頻率，男生女生都差沒有多少，停留時間男生跟女生大部分也都相同，所有男女生兩者的差距並不明顯。</span>**

## 問題8 店內氣氛評價最高和最低的分別是哪些顧客群?原因為何?

```python
rating_groups = df.groupby("氣氛評價")

for rating, group in rating_groups:
    print(f"氣氛評價為 {rating} 的顧客群:")
    print(group["職業"].value_counts())
    print(group["年齡"].value_counts())
```

```python
rating_groups = df.groupby("氣氛評價")

plt.figure(figsize=(12, 6))
for rating, group in rating_groups:
    ax = group["職業"].value_counts().plot(kind='bar', label=f'氣氛評價={rating}')
    ax.set_xlabel('職業')
    ax.set_ylabel('數量')
plt.title('不同氣氛評價分數的顧客職業分布')
plt.legend()
plt.show()

plt.figure(figsize=(12, 6))
for rating, group in rating_groups:
    ax = group["年齡"].value_counts().plot(kind='bar', label=f'氣氛評價={rating}')
    ax.set_xlabel('年齡')
    ax.set_ylabel('數量')  
plt.title('不同氣氛評價分數的顧客年齡分布')
plt.legend()
plt.show()
```

## 問題9 WiFi品質得到不同評價的顧客,他們其他評價的差異為何?

```python
wifi_groups = df.groupby("WiFi品質評價")

for rating, group in wifi_groups:
    print(f"WiFi品質評價為 {rating} 時的其他評價:")
    print(group[["產品評價", "價格評價", "促銷評價", "氣氛評價", "服務評價", "選擇評價"]].mean())
```
